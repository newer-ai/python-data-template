```python
import matplotlib.pyplot as plt
import numpy as np

from utils.plot_ext import config_plt

config_plt()

# 定义损失函数 L(w) = (w-2)^2 + 1 （一个简单的碗形函数）
w = np.linspace(-1, 5, 200)
L = (w - 2) ** 2 + 1

# 计算导数（斜率）
dL = 2 * (w - 2)

# 绘图
plt.figure(figsize=(8, 5))
plt.plot(w, L, label="损失函数 L(w)", color="blue")

# 标注最小值点
plt.scatter(2, 1, color="red", zorder=5)
plt.text(2, 1.2, "碗底 (最小值)", ha="center", color="red")

# 在左侧点画切线方向
w_point_left = 0.5
L_left = (w_point_left - 2) ** 2 + 1
slope_left = 2 * (w_point_left - 2)
plt.scatter(w_point_left, L_left, color="green")
plt.text(w_point_left, L_left + 1, "左侧点\n斜率<0\n往右走", ha="center", color="green")

# 在右侧点画切线方向
w_point_right = 3.5
L_right = (w_point_right - 2) ** 2 + 1
slope_right = 2 * (w_point_right - 2)
plt.scatter(w_point_right, L_right, color="orange")
plt.text(w_point_right, L_right + 1, "右侧点\n斜率>0\n往左走", ha="center", color="orange")

# 设置图像
plt.title("损失函数曲线与斜率方向示意")
plt.xlabel("参数 w")
plt.ylabel("损失 L(w)")
plt.grid(True)
plt.legend()
plt.show()
```
